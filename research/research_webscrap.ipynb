```python
from bs4 import BeautifulSoup
import requests
```

```python
URL = "https://shop.zuscoffee.com/collections/drinkware"
```

```python
page = requests.get(url=URL)
```

```python
soup = BeautifulSoup(page.content, "html.parser")
```

```python
content = soup.find_all(class_="product-card__title")
```

```python
content
```

```python
for idx, con in enumerate(content):
    print(idx)
    print(con.text)
```

```python
filter_1 = soup.find_all("div", class_="tab-content")
```

```python
for f in filter_1:
    print(f.text)
```
